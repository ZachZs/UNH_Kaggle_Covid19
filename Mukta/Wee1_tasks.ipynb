```python
# creating one df from all json using above logic


list = []

for i in range(0,4):
    for j in range(0,(len(os.listdir(Subfolders[i])))):
        jsonpath = Subfolders[i]+ '/' + os.listdir(Subfolders[i])[j]
        
        with open(jsonpath) as file:
            dict_json =  json.load(file)
            
            paper_id = dict_json['paper_id']
            title = dict_json['metadata']['title']
            abstract = ''
            for A_text in dict_json['abstract']:
                abstract = abstract + ' ' + A_text['text']
            body_text =''
            for B_text in dict_json['body_text']:
                body_text = body_text + ' ' + B_text['text']
            
            list.append({'paper_id':paper_id , 'title':title, 'abstract':abstract, 'body_text':body_text})
            

master_df = pd.DataFrame(list)
master_df
    

```
