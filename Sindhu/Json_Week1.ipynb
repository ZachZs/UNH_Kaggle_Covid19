```python
# FolderPath = "C:\Users\kumar\Documents\Analytics & Data Science_3rd semester\Mittal\COVID-19_Kaggle_Challenge"
CovidFolders = ["biorxiv_medrxiv/biorxiv_medrxiv", "comm_use_subset/comm_use_subset", 
              "custom_license/custom_license","noncomm_use_subset/noncomm_use_subset" ]
```

```python
# Reading all the json files using for loop
jsonfiles=[]
for i in CovidFolders:
    for j in os.scandir(i):
        with open(j) as f:
            jsonfile = json.load(f)
            jsonfiles.append(jsonfile)

df = pd.DataFrame(jsonfiles)
# print(df.columns)
df
```

```python
# etract data from each json file using for loop
from datetime import datetime


start = datetime.now()

files=[]
for i in CovidFolders:
    for j in os.scandir(i): 
        with open(j) as f:
            jsonfile = json.load(f)            

            paper_id = jsonfile["paper_id"]
            title = jsonfile["metadata"]["title"]
            
            abstract = ""
            for ab in jsonfile["abstract"]:
                abstract = abstract + "" + ab["text"]           
            body = ""
            for bo in jsonfile["body_text"]:
                body = body + "" + bo["text"]
    
            files.append({"PaperID": paper_id, "Title": title, "Abstract": abstract, "Body":body })

Main_DF = pd.DataFrame(files)
end=datetime.now()
print("Run time: ", end-start)

Main_DF
```
