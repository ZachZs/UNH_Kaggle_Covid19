```python
# -----------------------------------------------------------
# Importing all the file names into lists
# -----------------------------------------------------------



%cd "C:\Users\Jordan\Desktop\COVID-19 Challenge"

import os

bio_med_files = []
comm_files = []
custom_files = []
noncomm_files = []

for f in os.listdir("biorxiv_medrxiv"):
    bio_med_files.append(f)
for f in os.listdir("comm_use_subset"):
    comm_files.append(f)
for f in os.listdir("custom_license"):
    custom_files.append(f)
for f in os.listdir("noncomm_use_subset"):
    noncomm_files.append(f)
    
```

```python
## Lengths of files match up to readme and directory count
print("Length of {}: {} ".format("bio_med_files", len(bio_med_files)))
print("Length of {}: {} ".format("comm_files", len(comm_files)))
print("Length of {}: {} ".format("custom_files", len(custom_files)))
print("Length of {}: {} ".format("noncomm_files", len(noncomm_files)))
```

```python
# -----------------------------------------------------------
# Creating lists for the different key value pairs in the json dictionary
# -----------------------------------------------------------

##defining function so that I don't repeat myself

## We also only want the paper id, title, abstract, and actual text of the paper
def json_reader(r):
    read_file = open(r,)
    data = json.load(read_file)  ## data is a dictionary
    paper_id.append(data['paper_id'])
    title.append(data['metadata']['title'])
    
    abstract_str = ''
    for text in data['abstract']:
        abstract_str = abstract_str + ' ' + text['text'] 

    body_text_str = ''
    for text in data['body_text']:
        body_text_str = body_text_str + ' ' + text['text']

    abstract.append(abstract_str)
    body_text.append(body_text_str)

    read_file.close()
    return
    
paper_id = []
title = []
abstract = []
body_text = []

file_location = []


print(datetime.datetime.now())

%cd "C:\Users\Jordan\Desktop\COVID-19 Challenge\biorxiv_medrxiv"
for r in bio_med_files:
    json_reader(r)
    file_location.append("biorxiv_medrxiv")

print(datetime.datetime.now())
%cd "C:\Users\Jordan\Desktop\COVID-19 Challenge\comm_use_subset"
for r in comm_files:
    json_reader(r)
    file_location.append('comm_use subset')

print(datetime.datetime.now())
%cd "C:\Users\Jordan\Desktop\COVID-19 Challenge\custom_license"
for r in custom_files:
    json_reader(r)
    file_location.append("custom_license")

print(datetime.datetime.now())
%cd "C:\Users\Jordan\Desktop\COVID-19 Challenge\noncomm_use_subset"
for r in noncomm_files:
    json_reader(r)
    file_location.append("noncomm_use_subset")

print(datetime.datetime.now())
```
