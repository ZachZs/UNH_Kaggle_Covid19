### 3. Data processing step (Tokenization ,Remove punctuation and special characters ,lowercase)

```python
import nltk
nltk.download('stopwords')
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
import pandas as pd
import numpy as np
```

```python
tokenizer = RegexpTokenizer(r'\w+')
stemmer = PorterStemmer() 
stop_words = set(stopwords.words('english'))
```

```python
# excel = pd.read_csv(open(df, 'rb')) #use pandas to read excel file
strdata = extract_data.values.tolist() #convert values to list of lists (each row becomes a separate list)
print(len(strdata))

tokens = [tokenizer.tokenize(str(i)) for i in strdata] #tokenize words in lists
print(tokens)
```

```python

cleaned_list = []

for w in tokens:
    stopped = [i for i in w if str(i).lower() not in stop_words] #remove stop words
    stemmed = [stemmer.stem(i) for i in stopped] #stem words
    cleaned_list.append(stemmed) #append stemmed words to list
```

### 4. The objective is to create a Search engine to find papers based on key words.

Use Regular expressions to find one word

A combination of words (do it on body text)


```python
input_word = input("search:")

stop_word = set(stopword.words('english'))
input_word = input_word.lower().split(" ")
input_word = word for word in input_word if word not in stop_word


word_count = []
drops = []

#input pattern 
pattern = ''
for word in input_word:
    pattern = pattern +word 
    
#search whole document

for i,j in enumerate(cleaned_list):
    j = j.lower()
#     match =re.findall("{}".format(pattern), j)
    match = re.match(pattern, j)
    z = 0
    for word in match:
         if word != '': z+=1
    if z > 0:
        word_count.append(z)
        drops.append(True)
    else:
        drops.append(False)


final_df = extract_data[drops]
final_df['word_counts'] = word_count 
```
