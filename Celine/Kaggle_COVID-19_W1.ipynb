```python
start = datetime.now()

all_jsonfiles=[]
for folders in subfolders:
    for filename in os.scandir(folders): #"os.scandir" give path not just filename in that path
        with open(filename) as f:
            jsonfile = json.load(f)            

            paper_id = jsonfile["paper_id"]
            
            title = jsonfile["metadata"]["title"]
            
            abstract_text = ""
            for a_text in jsonfile["abstract"]:
                abstract_text = abstract_text + "" + a_text["text"]
                    
            body_text = ""
            for b_text in jsonfile["body_text"]:
                body_text = body_text + "" + b_text["text"]
    
            all_jsonfiles.append({"paper_id": paper_id, "title": title, "abstract": abstract_text, "body_text":body_text })

df_all = pd.DataFrame(all_jsonfiles)


end=datetime.now()
print(start)
print(end)
print("Run time: ", end-start)

df_all.head()
```
